# Correction exercice pandas

## Objectif de l'exercice
- Réaliser une visualisation de la taille de l'abstract des articles qui contiennent "social" dans leur titre

## Décomposition
- filtrer les articles qui contiennent python dans leur titre
- calculer la taille de l'abstract
- réaliser un histogramme

**Des aides et idées :**
- .dropna(subset="mar_var")
- la fonction str.lower()
- la fonction str.contains()
- la fonction len() ou la fonction str.len()
- .plot(kind="hist")

```python
# Charger pandas
import pandas as pd

# Charger les données
data_url = "https://raw.githubusercontent.com/pyshs/CUSO2026/refs/heads/main/data/css_openalex_26022026.csv"
df = pd.read_csv(data_url)
```

```python
# Virer les lignes sans titre
df = df.dropna(subset="title")
# créer un filtre vérifiant si le titre contient "social" (en minuscule)
mask = df["title"].str.lower().str.contains("social")
# appliquer le filtre
df = df[mask]
```

```python
# Créer une fonction pour compter la taille de l'abstract
def compter(texte):
    if pd.isnull(texte):
        return None
    else:
        return len(texte)


# Utiliser .apply et créer une nouvelle variable pour la taille de l'abstract
#
df["len_abstract"] = df["abstract"].apply(compter)
```

```python
# voir ce que ça donne
df["len_abstract"].describe()
```

```python
# en fait, je pourrais utiliser directement .str.len()
df["abstract"].str.len().describe()
```

```python
# Faire un histogramme basique
df["len_abstract"].plot(kind="hist")
```

```python
# fignoler un peu le graph :
df["len_abstract"].plot(
    kind="hist",  # le type de graph
    bins=50,  # le nombre de bin
    title="Distribution de la taille des abstracts",  # un titre
    xlim=(0, 10000),  # limiter à l'axe des x et couper à 10000
)
```
