## Lancer un script

`````python
# Calcule et affiche la proportion de mots de plus de 4 lettres d'une phrase
nombre_min_lettres = 4 # créer une variable
sortie_fichier = True # créer une variable
print("Calcul de la proportion de mots avec 4 lettres ou plus.") # afficher un texte à l'écran
phrase = input("Écrivez une phrase: ") # récupérer une entrée de l'utilisateur dans une varioable
liste_mots = phrase.split(" ") # on découpe la chaine de caractères aux espaces
nombre_mots = len(liste_mots) # on renvoie la longueur de la liste
if len(phrase) > 0: # on teste si l'utilisater a écrit quelque chose
    compteur = 0
    for i in liste_mots:
        if len(i) >= nombre_min_lettres:
            compteur += 1
    proportion = round(100 * compteur / nombre_mots, 2)
    informations = {
    "Total": compteur,
    "Proportion": proportion,
    "Phrase": phrase,
    "Seuil": nombre_min_lettres,
    }
    sortie = (
        f"Proportion de mots avec {nombre_min_lettres} lettres ou plus : {proportion}"
    )
    print(sortie)
    if sortie_fichier:
        with open("resultat.txt", "w") as f:
            f.write(str(informations))
else:
    print("Vous n'avez rien écrit.")
`````
