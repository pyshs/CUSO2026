# Exercice manipulation de données avec pandas

## Objectif de l'exercice
- Réaliser une visualisation de la taille de l'abstract des articles qui contiennent "social" dans leur titre

## Décomposition
- charger les données
- filtrer les articles qui contiennent python dans leur titre
- calculer la taille de l'abstract
- réaliser un histogramme

**Des aides et idées :**
- .dropna(subset="mar_var")
- la fonction str.lower()
- la fonction str.contains()
- la fonction len() ou la fonction str.len()
- .plot(kind="hist")
